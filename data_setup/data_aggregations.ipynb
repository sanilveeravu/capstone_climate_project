```python
import pandas as pd
```

```python
states_df = pd.read_csv("../data/lookup_data/states_with_lat_long.csv")
states_df
```

```python
tmax_df = pd.read_csv('../data/cleaned_data/tmaxdata.csv',names=["state_code", "date", "value"])
tmax_agg_df = tmax_df.groupby([tmax_df['state_code'],pd.to_datetime(tmax_df['date']).dt.strftime('%Y-%m')]).mean().reset_index()
tmax_agg_df.date = pd.to_datetime(tmax_agg_df.date)
tmax_agg_df
```

```python
tmin_df = pd.read_csv('../data/cleaned_data/tmindata.csv',names=["state_code", "date", "value"])
tmin_agg_df = tmin_df.groupby([tmin_df['state_code'],pd.to_datetime(tmin_df['date']).dt.strftime('%Y-%m')]).mean().reset_index()
tmin_agg_df.date = pd.to_datetime(tmin_agg_df.date)
tmin_agg_df
```

```python
prcp_df = pd.read_csv('../data/cleaned_data/prcpdata.csv',names=["state_code", "date", "value"])
prcp_agg_df = prcp_df.groupby([prcp_df['state_code'],pd.to_datetime(prcp_df['date']).dt.strftime('%Y-%m')]).mean().reset_index()
prcp_agg_df.date = pd.to_datetime(prcp_agg_df.date)
prcp_agg_df
```

```python
snow_df = pd.read_csv('../data/cleaned_data/snowdata.csv',names=["state_code", "date", "value"])
snow_agg_df = snow_df.groupby([snow_df['state_code'],pd.to_datetime(snow_df['date']).dt.strftime('%Y-%m')]).mean().reset_index()
snow_agg_df.date = pd.to_datetime(snow_agg_df.date)
snow_agg_df
```

```python
tmax_agg_df = pd.merge(tmax_agg_df, states_df, how='inner', on = 'state_code')
tmin_agg_df = pd.merge(tmin_agg_df, states_df, how='inner', on = 'state_code')
prcp_agg_df = pd.merge(prcp_agg_df, states_df, how='inner', on = 'state_code')
snow_agg_df = pd.merge(snow_agg_df, states_df, how='inner', on = 'state_code')
tmax_agg_df.to_csv("../data/cleaned_data/tmaxdata_avg_with_lat_long_monthly.csv", index=False)
tmin_agg_df.to_csv("../data/cleaned_data/tmindata_avg_with_lat_long_monthly.csv", index=False)
prcp_agg_df.to_csv("../data/cleaned_data/prcpdata_avg_with_lat_long_monthly.csv", index=False)
snow_agg_df.to_csv("../data/cleaned_data/snowdata_avg_with_lat_long_monthly.csv", index=False)
tmax_agg_df
```

```python
tmax_agg_yearly_df = tmax_df.groupby([tmax_df['state_code'],pd.to_datetime(tmax_df['date']).dt.strftime('%Y')]).mean().reset_index()
tmax_agg_yearly_df.date = pd.to_datetime(tmax_agg_yearly_df.date)
tmax_agg_yearly_df
```

```python
prcp_agg_yearly_df = prcp_df.groupby([prcp_df['state_code'],pd.to_datetime(prcp_df['date']).dt.strftime('%Y')]).mean().reset_index()
prcp_agg_yearly_df.date = pd.to_datetime(prcp_agg_yearly_df.date)
prcp_agg_yearly_df
```

```python
snow_agg_yearly_df = snow_df.groupby([snow_df['state_code'],pd.to_datetime(snow_df['date']).dt.strftime('%Y')]).mean().reset_index()
snow_agg_yearly_df.date = pd.to_datetime(snow_agg_yearly_df.date)
snow_agg_yearly_df
```

```python
tmax_yearly_df = tmax_df.groupby([tmax_df['state_code'],pd.to_datetime(tmax_df['date']).dt.strftime('%Y')])["value"].agg("max").reset_index()
tmax_yearly_df.date = pd.to_datetime(tmax_yearly_df.date)
tmax_yearly_df = pd.merge(tmax_yearly_df, states_df, how='inner', on = 'state_code')
tmax_yearly_df.to_csv("../data/cleaned_data/tmaxdata_max_with_lat_long_yearly.csv", index=False)
tmax_yearly_df
```

```python
tmin_yearly_df = tmin_df.groupby([tmin_df['state_code'],pd.to_datetime(tmin_df['date']).dt.strftime('%Y')])["value"].agg("min").reset_index()
tmin_yearly_df.date = pd.to_datetime(tmin_yearly_df.date)
tmin_yearly_df = pd.merge(tmin_yearly_df, states_df, how='inner', on = 'state_code')
tmin_yearly_df.to_csv("../data/cleaned_data/tmindata_min_with_lat_long_yearly.csv", index=False)
tmin_yearly_df
```

```python
tmax_yearly_df_1960=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='1961-01-01') & (tmax_agg_yearly_df.date<='1970-01-01')]
tmax_yearly_df_1960.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_1960=tmax_yearly_df_1960.groupby([tmax_yearly_df_1960.state_name]).mean()
tmax_yearly_df_1970=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='1971-01-01') & (tmax_agg_yearly_df.date<='1980-01-01')]
tmax_yearly_df_1970.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_1970=tmax_yearly_df_1970.groupby([tmax_yearly_df_1970.state_name]).mean()
tmax_yearly_df_1980=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='1981-01-01') & (tmax_agg_yearly_df.date<='1990-01-01')]
tmax_yearly_df_1980.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_1980=tmax_yearly_df_1980.groupby([tmax_yearly_df_1980.state_name]).mean()
tmax_yearly_df_1990=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='1991-01-01') & (tmax_agg_yearly_df.date<='2000-01-01')]
tmax_yearly_df_1990.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_1990=tmax_yearly_df_1990.groupby([tmax_yearly_df_1990.state_name]).mean()
tmax_yearly_df_2000=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='2001-01-01') & (tmax_agg_yearly_df.date<='2010-01-01')]
tmax_yearly_df_2000.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_2000=tmax_yearly_df_2000.groupby([tmax_yearly_df_2000.state_name]).mean()
tmax_yearly_df_2010=tmax_agg_yearly_df[(tmax_agg_yearly_df.date>='2010-01-01') & (tmax_agg_yearly_df.date<='2020-01-01')]
tmax_yearly_df_2010.drop(["latitude","longitude","date"],inplace=True,axis=1)
tmax_yearly_df_2010=tmax_yearly_df_2010.groupby([tmax_yearly_df_2010.state_name]).mean()
```

```python
tmax_avg_decade=tmax_yearly_df_1960.join(tmax_yearly_df_1970,on="state_name",how="inner",lsuffix='_1960',rsuffix='_1970')\
                                 .join(tmax_yearly_df_1980,on="state_name",how="inner")\
                                 .join(tmax_yearly_df_1990,on="state_name",how="inner",lsuffix='_1980',rsuffix='_1990')\
                                 .join(tmax_yearly_df_2000,on="state_name",how="inner")\
                                 .join(tmax_yearly_df_2010,on="state_name",how="inner",lsuffix='_2000',rsuffix='_2010')\

tmax_avg_decade
```

```python
tmax_variance=tmax_avg_decade
# tmax_variance["1960"]=(tmax_variance.value_2010-tmax_variance.value_1960)/tmax_variance.value_1960*100
# tmax_variance["1970"]=(tmax_variance.value_2010-tmax_variance.value_1970)/tmax_variance.value_1970*100
# tmax_variance["1980"]=(tmax_variance.value_2010-tmax_variance.value_1980)/tmax_variance.value_1980*100
# tmax_variance["1990"]=(tmax_variance.value_2010-tmax_variance.value_1990)/tmax_variance.value_1990*100
# tmax_variance["2000"]=(tmax_variance.value_2010-tmax_variance.value_2000)/tmax_variance.value_2000*100
tmax_variance["1960"]=(tmax_variance.value_2010-tmax_variance.value_1960)
tmax_variance["1970"]=(tmax_variance.value_2010-tmax_variance.value_1970)
tmax_variance["1980"]=(tmax_variance.value_2010-tmax_variance.value_1980)
tmax_variance["1990"]=(tmax_variance.value_2010-tmax_variance.value_1990)
tmax_variance["2000"]=(tmax_variance.value_2010-tmax_variance.value_2000)
tmax_variance.drop(["value_1960","value_1970","value_1980","value_1990","value_2000","value_2010"],inplace=True,axis=1)
tmax_variance
```

```python
tmax_variance=tmax_variance.reset_index().melt(id_vars='state_name')
tmax_variance.rename(columns = {'variable':'year'}, inplace = True)
tmax_variance["year"] = pd.to_numeric(tmax_variance["year"])
tmax_variance
```
