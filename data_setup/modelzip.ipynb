```python
import bz2
import glob
```

```python
modelFilesList = glob.glob('../model/gbr/modelfiles/*model.sav')
for modelFile in modelFilesList:
    print(modelFile)
    filename_in = modelFile
    filename_out = f"{modelFile}.bz2"
    with open(filename_in, mode="rb") as fin, bz2.open(filename_out, "wb") as fout:
        fout.write(fin.read())
```
